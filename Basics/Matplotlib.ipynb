```python
#matplotlib
#import matplotlib.pylot as plt
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
plt.plot([1,2,3],[4,5,1])
plt.title("Graph 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
style.use("ggplot")
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x,y,'g',label='Line 1',linewidth=2)
plt.plot(x2,y2,'b',label='Line 2',linewidth=2)
plt.title("Graph 2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(color='k')
```

```python
# plt.bar(x,y,
days1 =[0.25,1.25,2.25,3.25,4.25]
km1=[50,40,70,80,20]
plt.bar(days1,km1,
label="BMW",color='b',width=0.5)
days2 =[0.75,1.75,2.75,3.75,4.75]
km2 =[80,20,20,50,60]
plt.bar(days2,km2,
label="Audi",color='y',width=0.5)
plt.title("Graph 3")
plt.xlabel("Days")
plt.ylabel("Distance (in kms)")
plt.legend()
```
